- DataSet: NY Federal Reserve Credit Access Micro Data (https://www.newyorkfed.org/microeconomics/sce/credit-access#/)
- Goal: Predict Wheather a Credit Card Application will get Accepted(1) or Rejected(0).
- Total Columns: 135
- Columns Used: 34
- Discrete:19
- Continous:12
- Categorical:2
- Temporal:1
- Rows:5490 
- majority:minority = 85:15

- The Scoring metric I have chosen is ROC_AUC score, because for me reducing False Positive is important. The target variable I am trying to predict is wheather a credit card application will be granted or rejected. I want to reduce the error of classifying 1(Granted). i.e I don't want to predict that someones's application will be granted(1), when it will not in reality as this might affect applicant's credit score. And ROC_AUC tries to increase TP while reducing FP. 

- Best Model: 

- StackingClassifier With Logistic Regression as final model ({'final_estimator__C': 0.1, 'stack_method': 'auto'}), With the Following Models.


- 1.Extra Tree Classifier: {'criterion': 'gini', 'max_depth': 8, 'max_features': 'auto', 'n_estimators': 100}

```python
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

class log_add_const(BaseEstimator,TransformerMixin):
    def __init__(self,feature_names):
        self.feature_names = feature_names
        
    def fit(self,X,y = None):
        return self
    
    def transform(self,X,y = None):
        for var in self.feature_names:
            X[var] = np.log(X[var]+0.001)
        return X
```
